```python
import pandas as pd

# Load dataset
df = pd.read_csv("loan_data_sample 3.csv")  # modify path if needed

# Select only numerical columns
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

outlier_summary = []

for col in numeric_cols:
    # Calculate IQR
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    # Outlier thresholds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Count outliers
    outlier_count = len(df[(df[col] < lower_bound) | (df[col] > upper_bound)])

    outlier_summary.append({
        "Column": col,
        "Outlier_Count": outlier_count,
        "Total_Rows": len(df),
        "Outlier_%": round((outlier_count / len(df)) * 100, 2)
    })

# Convert to DataFrame
outlier_report = pd.DataFrame(outlier_summary)

# Filter columns where outliers exist
cols_with_outliers = outlier_report[outlier_report["Outlier_Count"] > 0]["Column"].tolist()

# Output
print("\n=== Outlier Report ===")
print(outlier_report.sort_values(by="Outlier_Count", ascending=False))

print("\n=== Columns With Outliers ===")
print(cols_with_outliers)

print(f"\nTotal Columns With Outliers: {len(cols_with_outliers)}")
```
